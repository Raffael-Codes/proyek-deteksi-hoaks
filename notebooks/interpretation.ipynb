```python
import joblib

print("--- Memuat Otak AI yang Sudah Dilatih ---")
vectorizer = joblib.load('../models/tfidf_vectorizer.pkl')
model = joblib.load('../models/logreg_model.pkl')
print("Model berhasil dimuat dan siap digunakan!")

print("\n--- Uji Coba Manual (Interpretasi) ---")
# Mari kita buat contoh berita palsu dan berita asli
berita_baru = [
    "Pemerintah akan membagikan uang tunai 10 juta untuk setiap warga negara secara gratis bulan ini! Klik link berikut.",
    "KPU telah menetapkan jadwal resmi pemilihan umum presiden tahun depan sesuai dengan undang-undang yang berlaku."
]

# Mengubah teks ke angka (TF-IDF) lalu menebaknya
berita_vektor = vectorizer.transform(berita_baru)
prediksi = model.predict(berita_vektor)

for teks, hasil in zip(berita_baru, prediksi):
    print(f"\nTeks Berita: '{teks}'")
    print(f"Prediksi AI: ---> **{hasil.upper()}**")
```
