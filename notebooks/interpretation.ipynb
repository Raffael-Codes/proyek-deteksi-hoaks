```python
import joblib
import sys
import os

# --- KUNCI UTAMA: Memberi tahu Python letak folder 'src' ---
# Mundur satu direktori ('..'), lalu masuk ke folder 'src'
path_ke_src = os.path.abspath(os.path.join('..', 'src'))
if path_ke_src not in sys.path:
    sys.path.append(path_ke_src)

# Sekarang Python sudah tahu tempatnya, kita bisa mengimpor fungsinya!
from data_preprocessing import bersihkan_teks 
# -----------------------------------------------------------

print("--- Memuat Otak AI yang Sudah Dilatih ---")
vectorizer = joblib.load('../models/tfidf_vectorizer.pkl')
model = joblib.load('../models/logreg_model.pkl')
print("Model berhasil dimuat dan siap digunakan!")

print("\n--- Uji Coba Manual (Interpretasi) ---")
# Contoh berita palsu dan asli bertema politik
berita_baru = [
    "Beredar pesan berantai di WhatsApp yang menyebutkan KPU terbukti melakukan kecurangan dengan membakar jutaan kotak suara! Tolong viralkan!",
    "KPU telah menetapkan jadwal resmi pemilihan umum presiden tahun depan sesuai dengan undang-undang yang berlaku."
]

# 1. BERSIHKAN TEKS DULU (Agar sama persis dengan kondisi saat model dilatih)
print("Sedang membersihkan teks input...")
berita_bersih = [bersihkan_teks(teks) for teks in berita_baru]

# 2. Mengubah teks BERSIH ke angka (TF-IDF) lalu menebaknya
berita_vektor = vectorizer.transform(berita_bersih)
prediksi = model.predict(berita_vektor)

for teks, hasil in zip(berita_baru, prediksi):
    print(f"\nTeks Berita: '{teks}'")
    print(f"Prediksi AI: ---> **{hasil.upper()}**")
```
